```python
# Libraries Importing
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
import warnings 
warnings.filterwarnings('ignore')
```

```python
# Matches Table
matches = pd.read_csv(r"C:\Users\BHUSHAN\Downloads\IPL Dataset 2008-24\matches.csv")
matches.head()
```

```python
# Updating Team Duplicates Name 
team_name_mappings = {
    "Deccan Chargers": "Delhi Capitals",
    "Delhi Daredevils": "Delhi Capitals",
    "Kings XI Punjab": "Punjab Kings",
    "Punjab Kings": "Punjab Kings",
    "Rising Pune Supergiants": "Rising Pune Supergiant",
    "Royal Challengers Bengaluru": "Royal Challengers Bangalore"
}

columns_to_update = ['team1', 'team2', 'toss_winner', 'winner']

for column in columns_to_update:
    matches[column] = matches[column].replace(team_name_mappings)
    
matches.head()
```

### 1. Distribution of match results

```python
plt.figure(figsize=(10, 4))
ax = sns.countplot(data=matches, x='result')
for bars in ax.containers:
    ax.bar_label(bars)
plt.title('Distribution of Match Results')
plt.show()
```

### 2. Distribution of runs scored by year-wise

```python
# Extract the year from the date
matches['year'] = matches['date'].dt.year

Both_t = deliveries.merge(matches, left_on='match_id', right_on='id', how='left')

yearly_runs = Both_t.groupby('year')['total_runs'].sum()

plt.figure(figsize=(12, 4))
ax = sns.barplot(x=yearly_runs.index, y=yearly_runs.values, palette="YlOrBr")
plt.title('Total Runs Scored Year-wise in IPL')
plt.xlabel('Year')
plt.ylabel('Total Runs')
plt.xticks(rotation=45)
for bars in ax.containers:
    ax.bar_label(bars)
plt.show()
```

### 3. Distribution of wickets taken year-wise

```python
yearly_wickets = Both_t.groupby('year')['is_wicket'].sum()

plt.figure(figsize=(12, 4))
ax = sns.barplot(x=yearly_wickets.index, y=yearly_wickets.values, palette="Blues")
plt.title('Total Wickets Year-wise in IPL')
plt.xlabel('Year')
plt.ylabel('Total Wickets')
plt.xticks(rotation=45)
for bars in ax.containers:
    ax.bar_label(bars)
plt.show()
```

### 5. Most Knockout Matches Appearence By Teams

```python
knockout_matches = matches[matches['match_type'].isin(["Final", "Semi Final", "Qualifier 1", "Qualifier 2", "Eliminator"])]
```

```python

team1_counts = knockout_matches['team1'].value_counts()
team2_counts = knockout_matches['team2'].value_counts()
winner = knockout_matches['winner'].value_counts()

# Combine the counts
total_counts = team1_counts.add(team2_counts, fill_value=0).astype(int)

plt.figure(figsize=(10,4))
result = pd.concat([total_counts, winner], axis=1, keys=['Total Knockut Matches', 'Wins'])
ax = result.sort_values(by="Total Knockut Matches", ascending=False).fillna(0).astype('int').plot(kind="bar")

for bars in ax.containers:
    ax.bar_label(bars)
    plt.title("Knockout Matches Most Appearance Or Win Matches In Konockouts")
plt.show()
```

### 6. Highest Win Percentage In Knockout Matches

```python
result = pd.concat([total_counts, winner], axis=1, keys=['Total Knockut Matches', 'Wins']).fillna(0).astype('int')
result["win_percetage"] = result.Wins/result['Total Knockut Matches']*100
result.sort_values(by="win_percetage", ascending=False).head(10).win_percetage.plot(kind="pie", autopct='%1.1f%%', colors=sns.color_palette("Paired"))
plt.ylabel("")
plt.title('Highest Win Percentage In Knockout Matches')
plt.show()
```

## Match Outcomes

### 20. Avg Inning Score For Each Venue 1st Or 2nd Inning

```python
# Merge matches and deliveries on match_id
merged_df = pd.merge(deliveries, matches, left_on='match_id', right_on='id')

# Calculate total runs for each inning at each venue
venue_inning_runs = merged_df.groupby(['venue', 'inning'])['total_runs'].sum().reset_index()

# Calculate the number of matches played at each venue
matches_per_venue = matches['venue'].value_counts().reset_index()
matches_per_venue.columns = ['venue', 'matches_played']

# Filter venues with more than 23 matches
matches_per_venue = matches_per_venue[matches_per_venue.matches_played > 23]

# Merge venue_inning_runs with matches_per_venue to calculate average scores
merged_df = pd.merge(venue_inning_runs, matches_per_venue, on='venue')
merged_df['avg'] = merged_df['total_runs'] / merged_df['matches_played']

# Select relevant columns and pivot the table
avg_score = merged_df[['venue', 'inning', 'avg']]
reshaped_df = avg_score.pivot(index='venue', columns='inning', values='avg').reset_index()
reshaped_df.columns = ['Venue', '1st Inning Avg Score', '2nd Inning Avg Score']

# Display the reshaped DataFrame
reshaped_df
```

### 21. Analyze the margins of wins by runs and wickets.

```python

# Filter out rows where the match was won by runs
wins_by_runs = matches[matches['result'] == 'runs']

# Filter out rows where the match was won by wickets
wins_by_wickets = matches[matches['result'] == 'wickets']

# Plot the distribution of win margins by runs
plt.figure(figsize=(12, 6))
sns.histplot(wins_by_runs['result_margin'], bins=30, kde=True)
plt.title('Distribution of Win Margins by Runs')
plt.xlabel('Run Margin')
plt.ylabel('Frequency')
plt.show()

# Plot the distribution of win margins by wickets
plt.figure(figsize=(12, 6))
sns.histplot(wins_by_wickets['result_margin'], bins=9, kde=True)
plt.title('Distribution of Win Margins by Wickets')
plt.xlabel('Wicket Margin')
plt.ylabel('Frequency')
plt.show()

```

### 22. Toss Impact on Match Results For Every Seasons

```python

# Analyze toss impact for each season
season_toss_impact = matches.groupby('season').apply(lambda x: (x[x['toss_winner'] == x['winner']].shape[0] / x.shape[0]) * 100).reset_index()
season_toss_impact.columns = ['Season', 'Toss Winning Impact']

print(season_toss_impact)

plt.figure(figsize=(12, 6))
sns.lineplot(data=season_toss_impact, x='Season', y='Toss Winning Impact', marker='o')
plt.title('Toss Winning Impact by Season')
plt.xlabel('Season')
plt.ylabel('Toss Winning Impact (%)')
plt.grid(True)
plt.show()
```
